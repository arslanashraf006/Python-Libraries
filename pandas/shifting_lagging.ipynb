```python
import pandas as pd

df = pd.read_csv('C:\\Users\\REGEN\\Documents\\python libraries\\pandas\\fb.csv', parse_dates=['Date'], index_col = 'Date')
df
```

```python
# shift the price to next days as shown in result
# as the shifted day now has no price so we add nan

df.shift(2)
```

```python
# now shifting to upward or right hand side like this
# shift method is also available on time series
# common use in finance is to calculate percentage change or price change in one day prices

df.shift(-1)
```

```python
df['Prev Day Price'] = df['Price'].shift(1)
df
```

```python
# now its easy to calculate price change in one day

df['1 day change'] = df['Price'] - df['Prev Day Price']
df
```

```python
# calculating 5 day return

df['5 day % return'] = (df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5)
df
```

```python
# keep data stir only want to shift dates for that we use this

df = df[['Price']]
df
```
