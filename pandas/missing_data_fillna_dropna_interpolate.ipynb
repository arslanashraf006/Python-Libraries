```python
# now convert day as my index value
df =  pd.read_csv("C:\\Users\\REGEN\\Documents\\python libraries\\pandas\\weather_data_missing.csv", parse_dates = ['day'])
df.set_index('day', inplace=True)
df
```

```python
# replace all nan values
```

```python
# fillna method for replace

new_df = df.fillna(0)
new_df
```

```python
#give different values to different data
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event',
})

new_df
```

```python
# convert values with forward fill method
# forward fill method replace the nan value to previous value in column
# new_df = df.fillna(method='ffill')
new_df = df.ffill()
new_df
```

```python
# backward fill method replace the nan value to next value in column
#new_df = df.fillna(method='bfill')
new_df = df.bfill()
new_df
```

```python
# axis method copying values horizantily
#new_df = df.fillna(method='bfill', axis='columns')
new_df = df.bfill(axis='columns')
new_df
```

```python
# limit the forward fill means limit the values to make how many copies to next empty values

new_df = df.ffill(limit = 1)
new_df
```

```python
# interpolate method gives the nan a value which is intermediate or middle between the cell values which at the start and end of the nan section

new_df = df.interpolate()
new_df
```

```python
# it changes the value near to the upcoming time value

new_df = df.interpolate(method='time')
new_df
```

```python
# drop na values method drop the rows which contains nan values

new_df = df.dropna()
new_df
```

```python
# drop row only if has all nan in data

new_df = df.dropna(how = "all")
new_df
```

```python
# if has given number of nan values than drop that row

new_df = df.dropna(thresh = 2)
new_df
```

```python
# insert the missing dates
dt = pd.date_range('01-01-2017', '01-11-2017')
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
