```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
# you can concate more data

df = pd.concat([india_weather,us_weather])
df
```

```python
# as the index is not continous to make it continous use ignore_index it will ignore original index

df = pd.concat([india_weather,us_weather], ignore_index = True)
df
```

```python
# want to retrive data of indian cities use keys for specific data

df = pd.concat([india_weather,us_weather], keys= ["india", "us"])
df
```

```python
# how you can retrive subset of data

df.loc['india']
```

```python
#if you want to concatinate data in new column wise for new key

#passing index is when you concatinate that it will concatinate the data in correct index with other data index

temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
# now this will add new column for windspeed

df = pd.concat([temperature_df, windspeed_df])
df
```

```python
# default axis is 0 means data is added in column wise if you change axis to 1 the data is added in row wise

df = pd.concat([temperature_df, windspeed_df], axis = 1)
df
```
