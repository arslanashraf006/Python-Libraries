```python
import pandas as pd

df = pd.read_excel("C:\\Users\\REGEN\\Documents\\python libraries\\pandas\\stock_data.xlsx", "Sheet1")
df
```

```python
# converion of cell content

def conver_people_cell(cell):
    if cell == "n.a.":
        return 'sam walton'
    return cell

def conver_eps_cell(cell):
    if cell == "not available":
        return None
    return cell
    
df = pd.read_excel("C:\\Users\\REGEN\\Documents\\python libraries\\pandas\\stock_data.xlsx", "Sheet1",
                  converters = {
                     'people': conver_people_cell,
                      'eps': conver_eps_cell
                  })
df
```

```python
# write to excel file

df.to_excel("C:\\Users\\REGEN\\Documents\\python libraries\\pandas\\new.xlsx", sheet_name = "stocks")
```

```python
# start writing at any point of file

df.to_excel("C:\\Users\\REGEN\\Documents\\python libraries\\pandas\\new.xlsx", sheet_name = "stocks", startrow =1, startcol=2,
           index = False)
```
