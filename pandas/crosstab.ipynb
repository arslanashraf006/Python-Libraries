```python
# cross tab
# In statistics a contingency table (also known as a cross tabulation or crosstab) is a type of table in a matrix format
# that displayes the frequency distribution of the variables

import pandas as pd
df = pd.read_excel("C:\\Users\\REGEN\\Documents\\python libraries\\pandas\\survey.xls")
df
```

```python
# first what you shows in your row and 2nd what you show in your column
# this will show the frequency of data means the number of occurence

pd.crosstab(df.Nationality, df.Handedness)
```

```python
pd.crosstab(df.Sex, df.Handedness)
```

```python
# to get the total of numbers use margins

pd.crosstab(df.Nationality, df.Handedness, margins=True)
```

```python
# multiple variables 
# I want to have nationality in column too

pd.crosstab(df.Sex, [df.Handedness, df.Nationality], margins = True)
```

```python
# multiple variables 
# I want to have nationality in row or x-axis too

pd.crosstab([df.Sex,  df.Nationality], [df.Handedness], margins = True)
```

```python
pd.crosstab(df.Nationality, df.Handedness, margins=True)
```

```python
#normalize gives you the percentage value

pd.crosstab(df.Nationality, df.Handedness, normalize='index')
```

```python
# want to know average age of people
import numpy as np

pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc = np.average)
```
