```python
# melt is used to transform or reshape data

import pandas as pd

df = pd.read_csv("C:\\Users\\REGEN\\Documents\\python libraries\\pandas\\weather_melt.csv")
df
```

```python
# id_vars are the things that you want on your column that you don't want to transform

df1 = pd.melt(df, id_vars=["day"])
df1
```

```python
# add filters

df1[df1["variable"]=="chicago"]
```

```python
# change the default column variable name

df1 = pd.melt(df, id_vars=["day"], var_name="city")
df1
```

```python
# change the default column value name

df1 = pd.melt(df, id_vars=["day"], var_name="city", value_name = "temperature")
df1
```
