```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:\\Users\\REGEN\\Documents\\python libraries\\pandas\\weather_data_replace.csv")
df
```

```python
# for single value change means if we have similar value all over
# new_df = df.replace(-99999, np.NaN)

#for multiple values to be changes
new_df =  df.replace([-99999, -88888], np.NaN)
new_df
```

```python
# replace based on column values

new_df = df.replace({
    "temperature": -99999,
    "windspeed": [-99999, -88888],
    "event": 0,
}, np.NaN)
new_df
```

```python
# mapping: replace values with new value

new_df = df.replace({
    -99999: np.NaN,
    'No Event': "Sunny"
})
new_df
```
