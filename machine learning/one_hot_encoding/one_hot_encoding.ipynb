```python
import pandas as pd
```

```python
df = pd.read_csv("C://Users//REGEN//Documents//python libraries//machine learning//one_hot_encoding//homeprices.csv")
df
```

```python
# In Pandas, the get_dummies() function converts categorical variables into dummy/indicator variables (known as one-hot encoding). 
# This method is especially useful when preparing data for machine learning algorithms that require numeric input.
dummies = pd.get_dummies(df.town, dtype = int)
dummies
```

```python
merged = pd.concat([df, dummies], axis = 'columns')
merged
```

Dummy Variable Trap


When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

NOTE: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library that you are using is not handling this for you

```python
# dummy variable trape-> read from google
# why we drop columns from dataframe
# sklearn linear model knows about one hot encoding and it drop all columns automatically you don't need to do these steps

final = merged.drop(['town', 'west windsor'], axis = 'columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop('price', axis='columns')
x
```

```python
y = final.price
y
```

```python
model.fit(x, y)
```

```python
# robinsville 2800 area price,0 for monroe township and 1 for robinsville
model.predict([[2800, 0, 1]])
```

```python
# west windsor 3400
model.predict([[3400, 0, 0]])
```

```python
# if you want to know how accurate your model is use score method
# score method calculate all the predict values from x and compare it with actual values with this y 
model.score(x,y)
```

```python
# Using sklearn OneHotEncoder

# First step is to use label encoder to convert town names into numbers

from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
# if you use the values it give you error not dataframe

X = dfle[['town','area']].values
X
```

```python
y = dfle.price.values
y
```

```python
# Now use one hot encoder to create dummy variables for each of the town

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
```

```python
model.predict([[1,0,2800]]) # 2800 sqr ft home in robbinsville
```
