```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("C://Users//REGEN//Documents//python libraries//machine learning//save_model_with_pickle_joblib//homeprices.csv")
df
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
# value of coffecient (means slope) from y = mx + b
reg.coef_
```

```python
# value of intercept

reg.intercept_
```

```python
# predict the price of area 5000 sqft

reg.predict([[5000]])
```

```python
# Save Model To a File Using Python Pickle

import pickle
```

```python
# Load Saved Model
# wb for writing also binary mode is important
with open('model_pickle', 'wb') as f:
    pickle.dump(reg,f)
```

```python
 # rb for reading also binary mode is important
with open('model_pickle', 'rb') as f:
     model = pickle.load(f)
```

```python
model.predict([[5000]])
```

```python
#Save Trained Model Using joblib
# if your model consist of large numpy array then using joblib might be more efficient

import joblib
```
