```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
plt.show()
```

```python
# how the line format appears in char specify with g is green color and then add format
plt.plot(x,y, 'g+')
plt.show()
```

```python
plt.plot(x,y, 'g+--')
plt.show()
```

```python
plt.plot(x,y, 'g+-.')
plt.show()
```

```python
# r is for red and the order does not matter how you add format and color
plt.plot(x,y, '--r')
plt.show()
```

```python
# D is for diamond
plt.plot(x,y, 'rD')
plt.show()
```

```python
# Has same effect like the uper code
plt.plot(x,y, color='red', marker='D', linestyle='')
plt.show()
```

```python
# for color you can also add hexadecimal format
plt.plot(x,y, color='#008840', marker='D', linestyle='')
plt.show()
```

```python
# you can also change the size of the marker with marker size
plt.plot(x,y, color='#008840', marker='D', linestyle='', markersize=20)
plt.show()
```

```python
# alpha will basically control the transparency
plt.plot(x,y, color='#008840', alpha = 0.2)
plt.show()
```
