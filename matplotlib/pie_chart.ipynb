```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Draw pie chart to track down home expenses

exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.show()
```

```python
# increase the radius of chart
# to show percentage value use the autopct
# if you want two decimal point then use 0.2 if one then use 0.1 if no then 0.0 etc
# explode use To "explode" a slice in a pie chart in Python
# if you want to start the pie from specific angle then use startangle then from example the home rent start from that angle like
plt.pie(exp_vals,labels=exp_labels, radius=2, autopct = '%0.2f%%', shadow=True, explode = [0,0.2,0,0,0], startangle=180)
plt.show()
```

```python
# how to save file in png
# bbox_inches is used to download the complete fit image
# pad_inches used to add extra padding 
# transparent is used to make the background of chart transparent
plt.savefig('piechart.png', bbox_inches="tight", pad_inches=2, transparent= True)
```
