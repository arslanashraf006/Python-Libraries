```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

# the label only shows when we call the legend property
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
# to change the legend location use the loc property by default it is 'best'
# to add shadow to legend use sadow property
plt.legend(loc="upper right", shadow=True, fontsize="small")
#to show grid in the chart use grid function
plt.grid()
plt.show()
```
